# Global Video Game Sales Forecasting

### Importing the Libraries and Data Files

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from bs4 import BeautifulSoup
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import pmdarima as pm
import statsmodels.api as sm
from pmdarima import auto_arima
from prophet import Prophet
from statsmodels.tsa.stattools import adfuller
```

```python
df = pd.read_csv(r"C:\Users\Jordan\Desktop\vgsales.csv")
```

```python
df = df.sort_values(by='Year').reset_index(drop=True)
df
```

# Exploratory Data Analysis ( EDA )

```python
print(df.info())
print(df.isnull().sum())
```

# Data Preprocessing

```python
df = df.dropna(subset=['Year'])
```

```python
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
```

```python
annual_sales = df.groupby(df['Year'].dt.year).agg({'Global_Sales': 'sum'}).reset_index()
```

```python
annual_sales.columns = ['Year', 'Global_Sales']
annual_sales.set_index('Year', inplace= True)
```

```python
annual_sales.index = pd.to_datetime(annual_sales.index.astype(str))
```

# Basic Vizualisation

```python
plt.figure(figsize=(10, 5))
plt.plot(annual_sales.index, annual_sales['Global_Sales'], marker='o')
plt.title('Global Video Game Sales Over Time')
plt.xlabel('Year')
plt.ylabel('Global Sales (in millions)')
plt.grid()
plt.show()
```

```python
# Check for NaN values
print(annual_sales.isnull().sum())
```

```python
# Differencing the series to make it stationary
annual_sales['Diff_Global_Sales'] = annual_sales['Global_Sales'].diff().dropna()

# Recheck stationarity on the differenced series
adf_result = adfuller(annual_sales['Diff_Global_Sales'].dropna())
print('ADF Statistic:', adf_result[0])
print('p-value:', adf_result[1])
```

# Auto ARIMA Model

```python
model = auto_arima(annual_sales['Diff_Global_Sales'].dropna(), seasonal=True, trace=True,
                   error_action='ignore', suppress_warnings=True)


print(model.summary())
```

```python
# Generate forecast for the next 5 periods (years) by first predicting on the differenced series
n_periods = 5
forecast_diff, conf_int_diff = model.predict(n_periods=n_periods, return_conf_int=True)

print("Forecast Differences:", forecast_diff)
```

```python
last_value = annual_sales['Global_Sales'].iloc[-1]

forecast = [last_value]

for diff in forecast_diff:
    last_value += diff
    forecast.append(last_value)

forecast_index = pd.date_range(start=annual_sales.index.max() + pd.DateOffset(years=1), periods=n_periods, freq='Y')

forecast_df = pd.DataFrame({
    'Forecasted': forecast[1:],
    'Lower Bound': conf_int_diff[:, 0] + annual_sales['Global_Sales'].iloc[-1],
    'Upper Bound': conf_int_diff[:, 1] + annual_sales['Global_Sales'].iloc[-1]
}, index=forecast_index)
```

## Auto-ARIMA Results

```python
# Plotting the results
plt.figure(figsize=(12, 6))
plt.plot(annual_sales['Global_Sales'], label='Historical Sales', color='blue')

# Plot forecasted values
if not forecast_df.empty:
    plt.plot(forecast_df.index, forecast_df['Forecasted'], label='Forecasted Sales', color='orange')
    plt.fill_between(forecast_df.index, 
                     forecast_df['Lower Bound'], 
                     forecast_df['Upper Bound'], 
                     color='lightgray', alpha=0.5, label='Confidence Interval')
else:
    print("No forecast data available.")

plt.title('Sales Forecast for the Next Year')
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.legend()
plt.grid()
plt.show()

# Print the forecasted values
print("Forecasted Values:")
print(forecast_df)
```

# PROPHET MODEL

```python
annual_sales = annual_sales.reset_index()
annual_sales = annual_sales[['Year','Global_Sales']]
```

```python
annual_sales.columns = ['ds', 'y']  # Rename columns for Prophet
annual_sales = annual_sales.sort_values(by='ds')

# Initialize the Prophet model
model = Prophet(yearly_seasonality=True)

# Fit the model
model.fit(annual_sales)

# Create a DataFrame to hold predictions
future = model.make_future_dataframe(periods=12, freq='Y')  # Forecast for the next 12 years

# Make predictions
forecast = model.predict(future)

# Plot the results
plt.figure(figsize=(12, 6))
model.plot(forecast)
plt.title('Forecast of Global Sales Using Prophet')
plt.xlabel('Year')
plt.ylabel('Global Sales (in millions)')
plt.grid()
plt.show()

# Print the forecasted values
print("Forecasted Values:")
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(12))
```
